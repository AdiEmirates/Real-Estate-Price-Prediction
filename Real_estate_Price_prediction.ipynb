# **Real Estate Price Prediction**



```
import numpy as np  # linear algebra
import pandas as pd # data processing, CSV file I/O
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```
df = pd.read_csv('/content/Real estate.csv')
df.head()
```

```
df.isnull().sum()
```

```
df.describe()
```

```
# Remove the 'No' column
df.drop('No', inplace = True, axis = 1)
```

```
# Rename feature names
df.columns = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'Y']
df.head()
```

```
# Info
print("Shape of dataframe: (m, n) =", df.shape)
df.info()
```

```
# Find correlation
df.corr()
```

```
plt.figure(figsize=(10,10))
sns.heatmap(data=df.corr(), annot=True, cmap='Greens')
```

```
# Remove the 'X1' column
df.drop('X1', inplace = True, axis = 1)
df.head()
```

```
# Data processing
x = df[['X2', 'X3', 'X4', 'X5', 'X6']]    # Features
y = df['Y']                               # Target

# Standardization (Z-score)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Split dataframe into Training_set and Testing_set
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```
print("Training_features set shape: x_train", x_train.shape)
print("Training_target   set shape: y_train", y_train.shape)

print("\nTesting_features set shape: x_test ", x_test.shape)
print("Testing_target   set shape: y_test ", y_test.shape)
```

```
# Linear regression
model = LinearRegression()
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)
```

```
# Performance
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}\n")
```

```
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid()
plt.plot([min(y_test),max(y_test)], [min(y_pred), max(y_pred)], color= 'red')
plt.title('Actual Price V/s Predicted Price')
```
